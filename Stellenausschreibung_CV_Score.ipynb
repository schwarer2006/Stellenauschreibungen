```python
# @title Standardtext für Titel
# Schritte 1 und 2: Daten laden und Textanalyse
# Annahme: Du verwendest die Natural Language Toolkit (NLTK) und Pandas

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Lade die Stellenausschreibung und den Lebenslauf
with open('/content/job_description.txt', 'r') as file:
    job_description = file.read()

with open('/content/resume.txt', 'r') as file:
    resume = file.read()

# Tokenisierung und Entfernen von Stoppwörtern
stop_words = set(stopwords.words('german'))

def preprocess_text(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
    return filtered_tokens

job_tokens = preprocess_text(job_description)
resume_tokens = preprocess_text(resume)

# Schritt 3: Vergleich und Matchen
# Annahme: Du verwendest einfache Heuristiken für den Vergleich, aber du könntest auch ML-Modelle verwenden

def match_keywords(tokens, keywords):
    matches = [token for token in tokens if token.lower() in keywords]
    return matches

job_keywords = ['business','sql', 'datenmodellierung', 'data analysis', 'kommunikation', 'sql server', 'agil', 'kompetenz', 'data warehouse','datenbank','java','deutsch', 'italienisch', 'sap', 'applikation']
resume_matches = match_keywords(resume_tokens, job_keywords)

# Schritt 4: Bewertung und Ausgabe
# Annahme: Einfache Bewertung basierend auf der Anzahl der Übereinstimmungen

matching_score = len(resume_matches) / len(job_keywords) * 100
resume_match   = len(resume_matches) * 10
job_match      = len(job_keywords) * 10

print(f"Matching Score: {matching_score}%")
print(f"Matching Resume: {resume_match}%")
print(f"Matching Resume: {job_match}%")
print(f"Matching Resume: {job_tokens}")
print(f"Matching Resume: {resume_tokens}")

# Sample text for demonstration
text = "'DEVOPS', 'BUSINESS', 'INTELLIGENCE', 'ENGINEER', 'Denner',  'Zürich', 'Zusammenarbeit', 'Entwicklungsteams', 'Fachbereichen', 'Integration', 'Daten', 'Datenanalysen', Anwendungen', 'Datenanbindung', 'Übernahme', 'SAP', 'Hana', 'Azure', 'Hilfe', 'Theobald', 'Extraktoren', 'Aufbau', 'neuen', 'Datenplattform', 'Datenmodellierung', 'Hilfe', 'WhereScape', 'Software', 'Verwaltung', 'Optimierung', 'Datenbanken', 'Data', 'Warehousing', 'Implementierung', 'Metadaten', 'getrieben', 'Dokumentationslösung', 'PowerBI', 'Cubes', 'Endanwender', 'Entwickler', 'SQL', 'Abfragen', 'Interne', 'Externe', 'Business', 'Partner', 'Pflege', 'Review', 'Azure', 'Wiki', 'Dokumentation', 'Monitoring', 'LadeJobs', 'Bearbeitung', 'Incidents', 'Anfragen', 'BUSINESS', 'INTELLIGENCE', 'INHOUSE', 'CONSULTANT', 'Bürkert', 'Fluid', 'Control', 'Systems', 'Ingelfingen', 'Deutschland', 'Erstellung', 'Beladen', 'Betreuung', 'multidimensionalen', 'Cubes', 'Microsoft', 'Analysis', 'Services', 'Identifikation', 'Datenquellen', 'Implementierung', 'Datentransformationen', 'Verbesserung', 'Datenqualität', 'Entwicklung', 'Implementierung', 'Unterstützung', 'Entscheidungsprozessen', 'DeltaMaster', 'Cubes', 'Schulung', 'Support', 'Mitarbeiter', 'Nutzung', 'Einführung', 'Microsoft', 'Reporting', 'Abteilungen', 'Sales', 'Pricing', 'Datenabfragen', 'Analysen', 'Fachbereiche', 'Erstellung', 'CustomerSignatur', 'analytischen', 'Auswertung', 'internen', 'Rabattstruktur', 'SSRS', 'Berichten', 'Applikationsbetreuung', 'Tagetik', 'Planungstool', 'Wolters', 'Kluwer', 'BI', 'Tools', 'DeltaMaster', 'Firma', 'Bissantz', 'Projekt', 'Mitarbeit', 'Datenlieferung', 'Produkt', 'Konfiguratoren', 'Rabattdaten', 'Projekt', 'Aufbau', 'Produktionsdaten', 'Cube', 'sowie', 'Bereitstellung', 'Daten', 'DeltaMaster', 'Userhelpdesk', 'Ticketbearbeitung', 'Agiles', 'Arbeiten', 'SCRUM', 'Schweiz', 'ZENTRALES', 'REPORTING', 'DATAWAREHOUSE', 'Diebold', 'Nixdorf', 'Tankstellen', 'Support', 'GmbH', 'Köln', 'Deutschland', 'Bereitstellen', 'Auswerten', 'Analysieren', 'prozessbezogenen', 'Kennzahlen', 'KPI', 'Bewertung', 'Geschäftsaktivitäten', 'Abstimmung', 'Fachbereichen', 'Datenabfrage', 'Datenanalyse', 'Aufbereitung', 'Kennzahlen', 'unterschiedliche', 'Fachbereiche', 'interne', 'externe', 'Businesspartner', 'Einhaltung', 'Umsetzung', 'Datenschutzes', 'gemäß', 'BSIDatenschutzrichtlinie', 'Personenbezogene', 'Daten', 'Durchführen', 'Dokumentieren', 'aktiven', 'bestehende', 'Reports', 'Änderungen', 'Datenstrukturen', 'Prozesse', 'Unterstützung', 'Fachbereiche', 'Erstellung', 'Analysen', 'Berichten', 'Mitwirken', 'fachspezifischen', 'Projekten', 'z', 'Einbindung', 'Datenquellen', 'Datawarehouse', 'mittels', 'ETL', 'Anbindung', 'Erstellung', 'Überwachung', 'ETL', 'Ladestrecken', 'SSIS', 'CONFIGURATION', 'MANAGEMENT', 'DATABASE', 'APPLICATIONS', 'Tankstellensupport', 'GmbH', 'Köln', 'Deutschland', 'Kenntnisse', 'Anpassung', 'Anwendungen', 'Anforderungen', 'Konfigurationsverwaltung', 'Eruierung', 'Analyse', 'Dokumentation', 'Bearbeitung', 'Fehlern', 'Änderungswünschen', 'Anforderungen', 'Optimierungen', 'Erfahrung', 'Zusammenarbeit', 'Abteilungen', 'Konfigurationsänderungen', 'koordinieren', 'verwalten', 'Fähigkeit', 'Analyse', 'Lösung', 'Problemen', 'Zusammenhang', 'Konfigurationsdaten', 'Unterstützung', 'Problembehebung', 'Fehlererkennung', 'Anwendungen', 'Basis', 'Konfigurationsdaten', 'Durchführung', 'regelmäßigen', 'Audits', 'Überprüfungen', 'Datenintegrität', 'Konfigurationsdatenbank', 'Abarbeitung', 'Erstellung', 'Reportanfragen', 'unterschiedlichen', 'Fachbereichen', 'Daten', 'DWH', 'Oracle', 'Datenbanken', 'ETL', 'Ladeprozesse', 'Erstellung', 'Hilfe', 'MS', 'Integration', 'Services', 'Interner', 'User', 'Support', 'Helpdesk', 'Software', 'Trainings', 'Stammdatenpflege', 'Betreuung', 'Verantwortlicher', 'September', 'MEHRSPRACHIGER', 'TECH', 'SUPPORT', 'Trainings', 'Anwednungsberatung', 'Tankstellensupport', 'GmbH', 'Köln', 'Deutschland', 'Erfahrung', 'Unterstützung', 'Kunden', 'technischen', 'Problemen', 'mehreren', 'Sprachen', 'First', 'SecondLevel', 'Support', 'Catch', 'Dispatch', 'Incidents', 'Durchführung', 'internationale', 'Kunden', 'Bereich', 'Maut', 'Kassensystemen', 'Kartenverarbeitung', 'Verantwortlich', 'Lösung', 'technischen', 'Herausforderungen', 'verschiedenen', 'Sprachen', 'Deutsch', 'Englisch', 'Französisch', 'Erfahrung', 'Umgang', 'Kunden', 'verschiedener', 'Nationalitäten', 'Kulturen', 'Ort', 'Kassensystem', 'Trainings', 'ARAL', 'Luxembourg', 'Erstellung', 'Bereitstellung', 'Schulungsmaterialien', 'Dokumenationen', 'technischen', 'Support', 'Interne', 'Projektmitarbeit', 'Einführung', 'BMC', 'Remedy', 'ARS', 'Organisation', 'Durchführung', 'Pächterwechselprozessen', 'ARAL', 'BP', 'Tankstellen', 'Anwendungsberatung', 'Warenwirtschaft', 'Juli', 'Mai', 'BOULANGER', 'PATISSIER', 'Diverse', 'Unternehmen', 'Luxemburg', 'Frankreich', 'unterschiedlichen', 'Unternehmen', 'Luxembourg', 'Frankreich', 'tätig', 'Aufgrund', 'Mehlstauballergie', 'konnte', 'Beruf', 'weitergführt', 'Ausbildung', 'Praktika', 'April', 'August', 'Praktikum', 'Bedodesign', 'Eisenach', 'Eisenach', 'Erfahrung', 'Bearbeitung', 'Verwaltung', 'multimedialem', 'Material', 'Kenntnisse', 'Gestaltung', 'Umsetzung', 'interaktiven', 'Teamarbeit', 'Produktion', 'audiovisuellen', 'Inhalten', 'verschiedene', 'Medien', 'Entwicklung', 'verschiedene', 'Plattformen', 'z', 'Webseiten', 'etc', 'Projekte', 'Filemaker', 'Datenbank', 'Juni', 'März', 'PC', 'ANWENDER', 'FACHMANN', 'Comdat', 'Institut', 'Computerschulung', 'Trier', 'Deutschland', 'Gesundheitsbedingte', 'Umorientierung', 'Mehlstauballergie', 'Umschulung', 'PC', 'Anwender', 'Fachmann', 'Klassischen', 'IT', 'Themen', 'Netzwerke', 'Betriebssysteme', 'Office', 'Programmierung', 'Pascal', 'Visual', 'Basisc', 'Datenbanken', 'MultiMedia', 'September', 'Juni', 'Ausbildung', 'Konditor', 'Bäcker', 'Boulangerie', 'Moulin', 'de', 'Berdorf', 'Echternach', 'Abschluss', 'Gesellenbrief', 'Zertifizierungen', 'Kurse', 'Business', 'Intelligence', 'MS', 'BI', 'Stack', 'GFU', 'SQL', 'SSAS', 'SSIS', 'SSRS', 'Zertifikate', 'DeltaMaster', 'Bissantz', 'Schulungen', 'OLAP', 'MDX', 'Kurs', 'inklusive', 'Zertifikat', 'Analytisches', 'SQL', 'Server', 'Udemy', 'Onlinetraining', 'SQL', 'Developer', 'GFU', 'IT', 'Kurse', 'allgemein', 'Office', 'Schulungen', 'Word', 'Excel', 'Access', 'Diverse', 'private', 'IT', 'Kurse', 'DOS', 'EXASOL', 'DB', 'Python', 'ITIL', 'Zertifizierung', 'GIT', 'Schulung', 'Zertifikate', 'Tagetik', 'Planungssoftware', 'Sonstige', 'Kurse', 'Betrieblicher', 'Ersthelfer', 'TMS', 'Kassensystem', 'Technikerschulung', 'Scheidt', 'Bachmann', 'Kompetenzen', 'Microsoft', 'BI', 'Stack', 'Datenanalyse', 'Teamfähigkeit', 'Datenvisualisierung', 'Datenmodellierung', 'DevOps', 'Azure', 'SQL', 'ETL', 'Tools', 'Hobbys', 'Freizeit', 'Sport', 'Radfahren', 'Schwimmen', 'Wandern', 'Kunst', 'Selber', 'Zeichnen', 'Abstrakt', 'Acryl', 'malen', 'Familie', 'Zeit', 'Familie', 'verbringen', 'Ausflüge', 'Brettspiele', 'Kino', 'Sonstiges', 'Bücher', 'lesen', 'Fachbuch', 'Krimi', 'Sprachen', 'Deutsch',  'Englisch', 'Fortgeschritten', 'Französisch', 'Fortgeschritten'"

#text = {job_tokens}

# Generate word cloud
wordcloud = WordCloud(width=899 ,height=699, background_color='white').generate(text)

# Display the generated word cloud using matplotlib
plt.figure(figsize=(8, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```
